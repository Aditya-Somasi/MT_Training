```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
ages = np.random.normal(20, 5, 1000)
ages
```

```python
mean = np.mean(ages)
print("Mean is", mean)
```

```python
median = np.median(ages)
print("Median is", median)
```

```python
mode_result = stats.mode(ages)
print("Mode is", mode_result)
```

```python
max_ages = max(ages)
min_ages = min(ages)
print("Max age is", max_ages)
print("Min age is", min_ages)
first_percentile = np.percentile(ages, 25)
print("25th percentile is", first_percentile)
third_percentile = np.percentile(ages, 75)
print("75th percentile is", third_percentile)
iqr = stats.iqr(ages)
print(iqr)
# print(third_percentile - first_percentile)
```

```python
import seaborn as sns
sns.histplot(ages,kde=True,bins=10)
plt.xlabel('ages')
plt.ylabel('frequency')
plt.title("distribution")
```

```python
lower_boundary = first_percentile - 1.5 * iqr
upper_boundary = third_percentile + 1.5 * iqr
print("Lower boundary is", lower_boundary)
print("Upper boundary is", upper_boundary)
outliers = ages[(ages < lower_boundary)|(ages > upper_boundary)]
print("Outliers are", outliers)
```

```python
filtered_ages = [float(age) for age in ages if lower_boundary <= age <= upper_boundary]
print("Filtered ages are", filtered_ages)
filtered_mean = np.mean(filtered_ages)
print("Mean after removing outliers is", filtered_mean)
print(len(filtered_ages))
```

```python
sns.histplot(filtered_ages,kde=True,bins=10)
plt.xlabel('ages')
plt.ylabel('frequency')
plt.title("distribution")
```
