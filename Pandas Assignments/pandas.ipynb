### Standardization (Z Score)

```python
mean_sales = sales_data.mean()
std_sales = sales_data.std()
z_score = (sales_data - mean_sales) / std_sales
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(sales_data, bins=30, color='skyblue')
plt.title('Original Sales Data')

plt.subplot(1,2,2)
plt.hist(z_score, bins=30, color='green')
plt.title('Z-score Standardized Sales Data')

plt.tight_layout()
plt.show()
```

```python
sales_data = data['Sales_Amount_is_win_NO']

# Filter out non-positive values before log transform
sales_data_pos = sales_data[sales_data > 0]
log_sales = np.log(sales_data_pos)

# Plot original vs log-transformed
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(sales_data_pos, bins=30, color='skyblue')
plt.title('Original Sales Data (Positive Only)')

plt.subplot(1,2,2)
plt.hist(log_sales, bins=30, color='orange')
plt.title('Log-transformed Sales Data')

plt.tight_layout()
plt.show()
```

### Handling outliers using +-3 method

```python
mask = (z_score >= -3) & (z_score <= 3)
filtered_sales = sales_data[mask]

# Print counts and first few rows
print(f"Original count: {len(sales_data)}")
print(f"Filtered count: {len(filtered_sales)}")
print(filtered_sales.head())
```

### Using Logarithamic transformation

```python
dropped_length = len(sales_data) - len(filtered_sales)
dropped_length
```

```python
log_filtered = np.log1p(filtered_sales)
```

```python
plt.subplot(1,2,2) 
plt.hist(log_filtered , bins=30, color='orange') 
plt.title('Log1p of Filtered Sales_Amount_is_win_NO') 
plt.tight_layout() 
plt.show()
```

### Churn

```python
from scipy.stats import f_oneway

sales_churn_0 = data[data['churn'] == 0]['Sales_Amount']
sales_churn_1 = data[data['churn'] == 1]['Sales_Amount']
anova_result = f_oneway(sales_churn_0, sales_churn_1)

# Step 4: Print results
print("ANOVA F-statistic:", anova_result.statistic)
print("ANOVA p-value:", anova_result.pvalue)

# Step 5: Interpretation
if anova_result.pvalue < 0.05:
    print("There is a statistically significant relationship between churn and sales amount.")
else:
    print("There is no statistically significant relationship between churn and sales amount.")
```
