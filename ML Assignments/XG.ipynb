```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
import pandas as pd
numeric_cols = df.select_dtypes(include='number').columns
scaler = StandardScaler()
df.loc[:, numeric_cols] = scaler.fit_transform(df[numeric_cols])
print(df.describe())

```

```python
#now select top 20 columns from the above tests
#based on p value we can select the top 20 columns
from sklearn.feature_selection import SelectKBest, f_classif
import pandas as pd

# Select only numerical features
numerical_cols = df.select_dtypes(include='number').columns.drop('churn', errors='ignore')

# Define X and y
X = df[numerical_cols]
y = df['churn']  # make sure 'churn' is your target column

# Apply SelectKBest with f_classif (ANOVA F-test for classification)
selector = SelectKBest(score_func=f_classif, k=20)
X_new = selector.fit_transform(X, y)

# Get the selected feature names
selected_mask = selector.get_support()
selected_features = numerical_cols[selected_mask]

# Print selected top 20 features
print("Top 20 numerical features selected:")
print(selected_features.tolist())

# Optional: Reduced DataFrame
X_selected = df[selected_features]
```

```python
from sklearn.feature_selection import SelectKBest, chi2

# Select only categorical columns (excluding target 'churn')
categorical_features = [col for col in categorical_cols if col != 'churn']
X_cat = df[categorical_features]
y_cat = df['churn']

# Convert categorical columns to numeric if needed (using get_dummies)
X_cat_encoded = pd.get_dummies(X_cat, drop_first=True)

# Apply SelectKBest with chi2 to select top 20 categorical features
selector_cat = SelectKBest(score_func=chi2, k=20)
X_selected_cat = selector_cat.fit_transform(X_cat_encoded, y_cat)

# Get selected feature names
selected_cat_mask = selector_cat.get_support()
selected_cat_features = X_cat_encoded.columns[selected_cat_mask]

print("Top 20 categorical features selected by chi-square:")
print(selected_cat_features.tolist())
```

```python
# Select only the columns in selected_cat_features from X_cat_encoded
X_selected_cat_df = X_cat_encoded[selected_cat_features]
print(X_selected_cat_df.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from xgboost import XGBClassifier
import pandas as pd

# Define the target variable
y = df['churn']

# Remove 'churn' from X_combined if it exists (safety check)
if 'churn' in X_combined.columns or 'churn_1.0' in X_combined.columns:
    X_combined = X_combined.drop(['churn', 'churn_1.0'], axis=1, errors='ignore')

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_combined, y, test_size=0.2, random_state=42, stratify=y
)

# Calculate scale_pos_weight to handle class imbalance
neg, pos = y_train.value_counts()
scale = neg / pos

# Initialize and train XGBoost model with class weight
model = XGBClassifier(
    eval_metric='logloss',
    random_state=42,
    scale_pos_weight=scale  # <-- Important change
)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print(" Accuracy Score:", accuracy_score(y_test, y_pred))
print(" Accuracy in Percentage:", accuracy_score(y_test, y_pred) * 100, "%")
print("\n Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\n Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Safety check to ensure target not in features
if 'churn' in X_combined.columns or 'churn_1.0' in X_combined.columns:
    X_combined = X_combined.drop(['churn', 'churn_1.0'], axis=1, errors='ignore')

# Target variable
y = df['churn']

# Train-test split (same as XGBoost)
X_train, X_test, y_train, y_test = train_test_split(
    X_combined, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
# Compute class weights
from sklearn.utils.class_weight import compute_class_weight

classes = np.unique(y_train)
class_weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
class_weight_dict = dict(zip(classes, class_weights))
print("Class weights:", class_weight_dict)
```

```python
# Initialize RandomForestClassifier
rf_model = RandomForestClassifier(
    n_estimators=200,        # number of trees
    random_state=42,
    class_weight=class_weight_dict,  # handle imbalance
    n_jobs=-1                # use all cores
)

# Train
rf_model.fit(X_train, y_train)

# Predict
y_pred = rf_model.predict(X_test)
```

```python
print(" Accuracy Score:", accuracy_score(y_test, y_pred))
print(" Accuracy in Percentage:", accuracy_score(y_test, y_pred) * 100, "%")
print("\n Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\n Classification Report:\n", classification_report(y_test, y_pred))
```

```python
importances = rf_model.feature_importances_
feat_imp = pd.Series(importances, index=X_combined.columns).sort_values(ascending=False)
print("Top features:\n", feat_imp.head(20))

# Plot top 20 features
plt.figure(figsize=(12, 6))
feat_imp.head(20).plot(kind='barh')
plt.title("Random Forest Feature Importance")
plt.gca().invert_yaxis()
plt.show()
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 0.5]
}

grid_rf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42, class_weight=class_weight_dict),
    param_grid=param_grid,
    scoring='f1',
    cv=5,
    n_jobs=-1,
    verbose=2
)

grid_rf.fit(X_train, y_train)
print("Best parameters:", grid_rf.best_params_)
```
