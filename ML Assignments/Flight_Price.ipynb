```python
## Separating date
data['Date'] = data['Date_of_Journey'].str.split('/').str[0]
data['Month'] = data['Date_of_Journey'].str.split('/').str[1]
data['Year'] = data['Date_of_Journey'].str.split('/').str[2]
```

```python
data['Date']=data['Date'].astype(int)
data['Month'] = data['Month'].astype(int)
data['Year'] = data['Year'].astype(int)
```

```python
data.head()
```

```python
data.drop('Date_of_Journey', axis = 1, inplace=True)
```

```python
data['Arrival_Time']= data['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
data['Arrival_Hour'] = data['Arrival_Time'].str.split(':').str[0]
data['Arrival_min'] = data['Arrival_Time'].str.split(':').str[1]
data
```

```python
data['Arrival_Hour']=data['Arrival_Hour'].astype(int)
data['Arrival_Hour'] = data['Arrival_Hour'].astype(int)
```

```python
data.drop('Arrival_Time', axis = 1, inplace = True)
```

```python
data
```

```python
data['Dep_hour'] = data['Dep_Time'].str.split(':').str[0]
data['Dep_min'] = data['Dep_Time'].str.split(':').str[1]
data.drop('Dep_Time', axis = 1, inplace = True)
```

```python
data['Dep_hour']=data['Dep_min'].astype(int)
data['Dep_min'] = data['Dep_min'].astype(int)
data
```

```python
# Handles cases where Duration may be like '2h 50m', '3h', or '50m'
def extract_hours(duration):
	if 'h' in duration:
		return int(duration.split('h')[0])
	else:
		return 0

def extract_minutes(duration):
	if 'm' in duration:
		return int(duration.split('m')[0].split()[-1])
	else:
		return 0

data['Duration_hours'] = data['Duration'].apply(extract_hours)
data['Duration_minutes'] = data['Duration'].apply(extract_minutes)
```
