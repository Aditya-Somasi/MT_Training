```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X, y = make_classification(n_samples=1000, n_redundant=0,n_features=2, n_clusters_per_class=1, weights=[0.90], random_state=12)
```

```python
X.shape, y.shape
```

```python
# Parameter Breakdown
# n_samples=1000: Creates 1000 data points (observations) in the dataset. This is the total number of rows that will be generated.​

# n_features=2: Generates only 2 features (columns) for each sample. This makes the dataset two-dimensional, which is ideal for plotting and visualizing the classification problem on a 2D scatter plot.​

# n_redundant=0: Ensures there are no redundant features created. Redundant features are linear combinations of informative features that don't add new information. Setting this to 0 means all features will be informative.​

# n_clusters_per_class=1: Creates exactly one cluster for each class. This means each class will have its data points grouped around a single centroid, making the classes more distinct and easier to separate.​

# weights=[0.90]: This creates a 90-10 class imbalance. The majority class will contain 90% of samples (900 samples), while the minority class will have only 10% (100 samples). This simulates real-world imbalanced datasets often encountered in fraud detection or rare disease prediction.​

# random_state=12: Sets the random seed to 12, ensuring reproducibility. Running this code multiple times will always generate the exact same dataset, which is crucial for consistent testing and comparison.​

# What Gets Generated
# The function returns two arrays: X with shape (1000, 2) containing the feature values, and y with shape (1000,) containing binary class labels (0 or 1). The resulting dataset will be simple to visualize since it only has 2 dimensions, with one tightly grouped cluster per class and a significant class imbalance favoring one class over the other.
```

```python
data1 = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
data2 = pd.DataFrame(y, columns =['Target'])
```

```python
final_data = pd.concat([data1, data2], axis=1)
final_data.head()
```

```python
final_data['Target'].value_counts()
```

```python
plt.scatter(final_data['Feature1'], final_data['Feature2'], c=final_data['Target'])
```

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
X_res, y_res = oversample.fit_resample(final_data[['Feature1', 'Feature2']], final_data['Target'])
```

```python
X.shape, y.shape
```

```python
X_res.shape, y_res.shape
```

```python
y_res==0
```

```python
data1 = pd.DataFrame(X_res, columns=['Feature1', 'Feature2'])
data2 = pd.DataFrame(y_res, columns =['Target'])
final_data1 = pd.concat([data1, data2], axis=1)
plt.scatter(final_data['Feature1'], final_data['Feature2'], c=final_data['Target'])
```

```python
## Before and after SMOTE
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.title('Before SMOTE')
plt.scatter(final_data['Feature1'], final_data['Feature2'], c=final_data['Target'])
plt.subplot(1, 2, 2)
plt.title('After SMOTE')
plt.scatter(final_data1['Feature1'], final_data1['Feature2'], c=final_data1['Target'])
```
